```python
import pandas as pd
df = pd.read_csv("./dataset_retriever.csv")
```

```python
from sentence_transformers import SentenceTransformer, losses, InputExample, models
from torch.utils.data import DataLoader
import pandas as pd
import random

# Charger le modèle de base
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')  # ou un autre

# Construction des exemples pour l'entraînement
train_examples = []
for _, row in df.iterrows():
    example = InputExample(texts=[row['query'], row['pertinent'], row['non_pertinent']])
    train_examples.append(example)

# Création du DataLoader
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=64)

# Perte contrastive par défaut (triplet loss)
train_loss = losses.TripletLoss(model=model)

# Fine-tuning
model.fit(
    train_objectives=[(train_dataloader, train_loss)],
    epochs=100,
    warmup_steps=100,
    output_path="output/fine_tuned_retriever7",
    use_amp=True,
)
```
